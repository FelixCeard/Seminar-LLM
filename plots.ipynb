```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
data = {
    'Orig': [23.5, 35.3, 52.9, 52.9],
    'Shuf': [14.1, 17.4, 36.1, 39.2],
    # 'shuf_std': [3.1, 3.5, 2.7, 3.9],
    'Trunc': [14.7, 17.6, 35.3, 32.4],
    'Shuf+Trunc': [8.1, 10.0, 22.1, 21.3],
    # 'shuf+trunc_std': [3.4, 3.0, 3.2, 3.7],
    'label': ['Bert_Base_k=1', 'Bert_Large_k=1', 'Bert_Base_k=5', 'Bert_Large_k=5' ]
}
df = pd.DataFrame(data)
df
```
